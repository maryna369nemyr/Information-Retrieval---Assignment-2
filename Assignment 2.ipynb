Let $X \in \mathbb{R}^ {p\times N},p< N$,be a centered matrix of $p$ $N$-dementional data samples with the full-size SVD $X=U\Sigma U^\top$, Assume that the singular values are sorted  in a descending manner, i.e .$\sigma_{1,1}\geq \dots \geq \sigma_{p,p}$.

 #### $\bullet $
 Provide a normalized vector $\hat{s}\in \mathbb{R}^p$, such that 
 
 $\DeclareMathOperator*{\argmin}{argmin}$
 ### $$\hat{s}= \mathop{\arg\max}_{s \ s.t. \ \|s\|=1}   s^\top \Sigma \Sigma ^\top s $$
 
 $\Sigma \Sigma^\top$ is a diagonal  $p \times p$ matrix, the diagonal entries of $\Sigma \Sigma^\top$ is $\sigma_{i,i}$, so $s^\top \Sigma \Sigma ^\top s$ can be written as the expansion form
 
### $$ s^\top \Sigma \Sigma ^\top s=\sum_{i=1}^p \sigma_{i}^{2} s_{i}^{2} $$

Morever, $\sigma_{1,1}$ is the largest entry, so 

### $$\sum_{i=1}^p \sigma_{i,i}^{2} s_{i}^{2} \leq \sigma_{1,1}^{2} \sum_{i=1}^p  s_{i}^{2}=\sigma_{1,1}^{2} $$

$\|s\|=1$,  the equality holds when $s=[1,0,0,\dots]^\top$
 
So, when $s=[1,0,0,\dots]^\top$, $ s^\top \Sigma \Sigma ^\top s$ has the maximum value $\sigma_{1,1}^{2}$

Let show that the empirical variance of the inner products of the colums of $\textbf{X}$ with a normalized vector $\textbf{a}$,

### $$

Assuming $\textbf{a}$ to be a linear combination of columns of $\textbf{U}$,  which form an orthonormal basis in ${\mathbb{R}^{p \times N}}$,  $\textbf{a}$ can be express as follows,

# Implementation of the k nearest neighbors

```python
from math import sqrt

def euclidean_distance(x1, x2):
    if(len(x1) != len(x2)):
        return 0.0;
    dist=0.0;
    for i in range(0, len(x1)):
        dist=dist + (x1[i] - x2[i])**2
    dist = sqrt(dist)
    return dist

def get_neighbors_labels(X_train, y_train, x_new, k):
    distances = []
    for i in range(0, len(X_train)):
        distances.append(euclidean_distance(x_new, X_train[i]))
        
    distances = np.array(distances)
    k_dist_ind = np.argsort(distances)[:k] # sorting distances and then returning their indencies
    return y_train[k_dist_ind]

#select the most common label among k nearest neighbors. 
def get_response(neighbors):
    from collections import Counter
    return Counter(neighbors).most_common(1)[0][0]

def compute_error(y_pred, y_test):
    prob = 0;
    for i in range(0,len(y_pred)):
        if(y_pred[i] != y_test[i]):
            prob+=1
    return(float(prob)/len(y_pred))

def predict(X_train, y_train, X_test, k):
    y_pred = []
    for x_new in X_test:
        neighbors = get_neighbors_labels(X_train, y_train, x_new, k)
        y_pred.append(get_response(neighbors))
    return y_pred
```

## Example of the KNN (k = 3) when we use 20 principal components and the training set is subset0.

The actual **reduced** (projected) data can be represented as 
$$P= AV = U\Sigma,$$ where its SVD is $$A = U\Sigma V^T,$$ in addition to the fact that the rows of the matrix $A$ are samples and columns are dimensions.

Pleas note, that if $$B= A^T \Rightarrow B = V\Sigma U^T ,$$
so $U$ would be an orthogonal matrix that would assure the projection and in this case it would be $B\cdot U$.


```python
# the actual projected/ reduced data can be represented as 
# P= AV = US, where A = USV^T and the rows of the matrix are samples, columns - dimensions

# reduced space, Vh_output_train includs only 20 singular values

#example of the error for 3NN when we use 20 principal components starting from the 1-st one.

begin_cut = 0
end_cut = 20

def singular_vectors_reduced(begin_cut, end_cut, X_train):
    U, S, Vh = np.linalg.svd(X_train, full_matrices=True)
    #print('Shapes of SVD U = {0}, S= {1}, Vh = {2}'.format(U.shape, S.shape,  Vh.shape))
    Vh_output_train = Vh[begin_cut:end_cut,:]
    return(Vh_output_train)

Vh_output_train = singular_vectors_reduced(begin_cut, end_cut, X_train)

X_train_reduced = np.matmul(X_train, (Vh_output_train.T))
X_test_reduced = np.matmul(X_test, (Vh_output_train.T))
print('Dimensions of the reduced training and test sets:',X_train_reduced.shape, X_test_reduced.shape)#print(y_train)

k_neighbours = 3
y_pred = predict(X_train_reduced, y_train, X_test_reduced, k_neighbours)
error = compute_error(y_pred, y_test)
print('Error = {0}'.format(error))
```

```python
# in addition to k - number of principle components I would also use 
# the number of the singular vector to start with (begin_cut)

def singular_vectors_reduced(begin_cut, end_cut, X_train):
    U, S, Vh = np.linalg.svd(X_train, full_matrices=True)
    #print('Shapes of SVD U = {0}, S= {1}, Vh = {2}'.format(U.shape, S.shape,  Vh.shape))
    Vh_output_train = Vh[begin_cut:end_cut,:]
    return(Vh_output_train)

def experiment(X_train, X_test, y_train, y_test, begin_cut, k_principal_components, k_neighbors):
    error = []
    for i in range(1,k_principal_components + 1):
        
        Vh_output_train = singular_vectors_reduced(begin_cut, begin_cut + i, X_train)
        #print('The shape of the matrix = {0}'.format(Vh_output_train.shape))
        X_train_reduced = np.matmul(X_train, (Vh_output_train.T))
        X_test_reduced = np.matmul(X_test, (Vh_output_train.T))
        y_pred = predict(X_train_reduced, y_train, X_test_reduced, k_neighbors)
        error.append(compute_error(y_pred, y_test))
    return(error)
        #print('Error = {0}'.format(error))
```

### The example of the proposed experiment regarding principal components, while the training set is subset0

```python
errors = experiment(X_train, X_test, y_train, y_test, 0, 20, 3)
errors_modif = experiment(X_train, X_test, y_train, y_test, 3, 17, 3)
```

```python
plt.plot(np.array(range(1,21)), errors, 'r', label = '1-st sing. value')
plt.plot(np.array(range(1,18)), errors_modif, 'b', label = '4-th sing. value')
plt.xlabel('# principal components')
plt.ylabel('error')
plt.title('The error (KNN)')
plt.grid()
plt.legend()
plt.show()

# plotting on since 4 -th singular vector gives us a better result (e.g. lower error) since v_4 is not the best fitting line
# for  a given column of X_train and its projection loses some information,
# and this lost information does not contribute to the error anymore 
```

Projecting on the principal components starting from 4-th singular vector gives us a better result (e.g. lower error) since $v_4$ is not the best fitting line for a given matrix $X_{train}$. 

Note: $$v_1  = \arg \max_{\lVert v \rVert_2 = 1}{\lVert X_{train} \cdot v \rVert_2}$$ is the best fitting line.

As a result, the projection of $X_{train}$ on the $span\{v_4, v_5, ...., v_{20}\}$ loses some information,
and this lost information does not contribute to the error anymore. 

For example, having two different vectors ${x_1, x_2}$ you can pick up a vector $v$ such that the projections of ${x_1, x_2}$ onto $v$ will result in points in the space with very small Euclidean distance between them, even though these vectors $x_1, x_2$ can be far away from each other.

### KNN when all the subsets play the role of a training set.

```python
## function over all train and test sets
def experiment_all_subsets(all_images, all_labels):
    dict_errors, dict_errors_modif = {}, {}
    for subset_name in (all_images):
        len_ = len(all_images[(subset_name)])
        print('Training set %s, number of images: %i' % (subset_name, len_)) #%e, %f
        # forming a training set 
        T = np.array(all_images[subset_name])
        X_train = np.reshape(T, (T.shape[0], -1))
        y_train = np.array(all_labels[subset_name])
        #forming a test set
        X_test, y_test = make_test_set(all_images, subset_name)
        X_test = np.reshape(X_test, (X_test.shape[0], -1))
        # the experiment with singular vectors
        print("Calculating the errors starting from the 1-st singular vector...")
        dict_errors[subset_name] = experiment(X_train, X_test, y_train, y_test, 0, 20, 3) 
        #0 - begin_cut ,e.g. from the first singular vector, 20 - max # of principal components, 3 - # of the nearest neighbors
        print("Calculating the errors starting from the 4-th singular vector...")
        dict_errors_modif[subset_name] = experiment(X_train, X_test, y_train, y_test, 3, 17, 3)
        #3 - begin_cut ,e.g. from the forth singular vector, 17 - max # of principal components, 3 - # of the nearest neighbors
    return(dict_errors, dict_errors_modif)
        
#dict_errors, dict_errors_modif = experiment_all_subsets(all_images, all_labels)
```
